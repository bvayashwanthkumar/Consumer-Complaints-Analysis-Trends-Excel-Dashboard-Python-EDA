# DATA-DRIVEN INSIGHTS INTO CONSUMER COMPLIANTS

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\BVAYA\Desktop\LPU\4th Semster\INT375 PYTHON\Project\consumerpython.csv")
```

## EDA Cell

```python
#Drop duplicates
data = data.drop_duplicates()

# First 10 rows
print("First 10 Rows")
print(data.head(10))

# Dataset shape
print("\nDataset Shape:", data.shape)

# Column names
print("\nColumns:")
print(data.columns.tolist())

# Data types
print("\nData Types:")
print(data.dtypes)

# Summary statistics for numeric columns
print("\nSummary Statistics (Numerical Columns):")
print(data.describe().info())

# Missing values
print("\nMissing Values:")
print(data.isnull().sum()[data.isnull().sum() > 0])
```

## Objective 1 :-  Compliant Volume By Product

```python
pd_c = data["Product"].value_counts()

# Mapping products to numbers
product_names = pd_c.index.tolist()
product_numbers = list(range(1, len(product_names) + 1))

# Creating bar plot with numeric x-axis
plt.figure(figsize=(12, 6))
sns.barplot(x=product_numbers, y=pd_c.values)

#Plotting
plt.yscale('log')
plt.xticks(product_numbers)  
plt.xlabel("Product Number")
plt.ylabel("Number of Complaints (Log Scale)")
plt.title("Complaint Count by Product (Log Scale)")

# Displaying product number mappings at the edge
print("\nðŸ“‹ Product Number Mapping:")
for i, name in enumerate(product_names, start=1):
    print(f"{i}: {name}")

print()
plt.tight_layout()
plt.show()
```

### Result Obtained
### -> Most complaints are related to checking and savings accounts.

### -> Credit cards and credit reporting services also receive many complaints.

### -> Only a few products generate the majority of complaints.

### -> Other products have very fewer complaints.

## Objective 2 :- Compliant Trends over Time

```python
# Converting 'Date received' to datetime
data['Date received'] = pd.to_datetime(data['Date received'], errors='coerce')

# Dropping rows with invalid or missing dates
data = data.dropna(subset=['Date received'])

# Groupping complaints by month
monthly_trend = data.groupby(data['Date received'].dt.to_period('M')).size()

# Convert index to datetime for plotting
monthly_trend.index = monthly_trend.index.to_timestamp()

# Plotting
plt.figure(figsize=(14, 6))
sns.lineplot(x=monthly_trend.index, y=monthly_trend.values, marker='o', linewidth=2)
plt.title("Complaint Trends Over Time", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Number of Complaints")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Result Obtained
### -> The number of consumer complaints increases over time.
### -> Complaint counts fluctuate from month to month.
### -> Higher complaint volumes are observed in recent years.
### -> Some months show sharp peaks in complaints..

-----------------------------------------------------------------------------------------------------------------------------------------------------

## Objective 3 :- Top 10 States with Most Compliant

```python
data = data.dropna(subset=['State'])

# Group by State and count complaints
state_complaints = data['State'].value_counts().head(10)  # Top 10 states
state_percentage=(state_complaints/state_complaints.sum())*100
#print("Percentage Contribution : ",percentage_contribution)

plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x=state_complaints.values,
    y=state_complaints.index,
    hue=state_complaints.index,
    palette='viridis'
)

# Add percentage labels inside the bars
for i, (count, pct) in enumerate(zip(state_complaints.values, state_percentage.values)):
    ax.text(
        count - (0.05 * count),      
        i,
        f"{pct:.1f}%",
        va='center',
        ha='right',
        color='white',
        fontsize=10,
        fontweight='bold'
    )

plt.title("Top 10 States with Most Complaints (Percentage Contribution)", fontsize=16)
plt.xlabel("Number of Complaints")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```

### Result Obtained
### -> California records the highest number of consumer complaints.
### -> Florida, Texas, and New York are the next major contributors.
### -> A small number of states contribute to the majority of complaints.
### -> Other states have comparatively lower complaint volumes.

## Objective 4 :- Compliant Submission Method Analysis

```python
# Count submission methods
method_counts = data['Submitted via'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x=method_counts.values,
    y=method_counts.index,
    hue=method_counts.index,
    palette='mako'
)

# Extend x-axis so labels are visible
ax.set_xlim(0, method_counts.max() * 1.15)

# Add exact values on bars
for bar in ax.patches:
    width = bar.get_width()
    y = bar.get_y() + bar.get_height() / 2
    ax.text(
        width + (0.01 * width),      
        y,
        f"{int(width)}",
        va='center',
        ha='left',
        fontsize=10
    )

#Labeling
plt.title("Complaint Submission Method Analysis", fontsize=16)
plt.xlabel("Number of Complaints")
plt.ylabel("Submission Method")
plt.tight_layout()
plt.show()
```

### Result Obtained
### -> Most consumer complaints are submitted through the web.
### -> Referral and phone are the next commonly used submission methods.
### -> Traditional methods such as postal mail, fax, and email are used very rarely.
### -> Consumers clearly prefer digital channels for submitting complaints.

## Objective 5 :- Timely vs Delayed Response

```python
#Unique Values
print("Unique values in 'Timely response?':")
print(data['Timely response?'].unique())
print()

# Count of responses
response_counts = data['Timely response?'].value_counts()
print("Count of Timely vs Delayed Responses:")
print(response_counts)
print()

# Percentage of each
response_percent = (response_counts / len(data)) * 100
print("Percentage of Timely vs Delayed Responses:")
print(response_percent.round(2))
print()

# Histogram (as a bar chart, since it's categorical)
plt.figure(figsize=(6, 4))
plt.bar(response_counts.index, response_counts.values, color=['skyblue', 'lightcoral'])
plt.title("Timely vs Delayed Responses")
plt.xlabel("Response Type")
plt.ylabel("Number of Complaints")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Result Obtained
### -> Most complaints received a timely response.
### -> A very small percentage of complaints were not responded to on time.
### -> The organization shows a high level of responsiveness to consumer complaints.

## Objective 6 :- Top 10 Sub Product Specific issues 

```python
# Grouping data by Sub-product and Issue
sub_product_issues = data.groupby(['Sub-product', 'Issue']).size().reset_index(name='Count')

# Display top 10 most frequent (Sub-product, Issue) pairs
top_issues = sub_product_issues.sort_values(by='Count', ascending=False).head(10)

# Create combined label
top_issues['Label'] = top_issues['Sub-product'] + " â€” " + top_issues['Issue']

# Plotting
plt.figure(figsize=(14, 7))
ax = sns.barplot(
    x='Count',
    y='Label',
    hue='Label',
    data=top_issues,
    palette='Blues_r'
)

# Adding value labels at end of bars
for bar in ax.patches:
    width = bar.get_width()
    y = bar.get_y() + bar.get_height() / 2
    ax.text(
        width + (0.01 * top_issues['Count'].max()),
        y,
        f"{int(width)}",
        va='center',
        ha='left',
        fontsize=10
    )

# Improve spacing
ax.set_xlim(0, top_issues['Count'].max() * 1.15)

# Labeling 
plt.title("Top 10 Sub-product Specific Issues", fontsize=16)
plt.xlabel("Number of Complaints")
plt.ylabel("Sub-product â€” Issue")
plt.tight_layout()
plt.show()
```

### Result Obtained
### -> Issues related to checking account management have the highest number of complaints.
### -> Credit card purchase problems and credit reporting errors are major complaint areas.
### -> A few sub-product and issue combinations account for most of the complaints.
### -> Other issues occur less frequently in comparison.
